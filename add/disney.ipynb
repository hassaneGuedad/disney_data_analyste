```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("disney.csv")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=data, palette='Set1')
plt.title('Count of TV Shows and Movies')
plt.xlabel('Type')
plt.ylabel('Count')
plt.gca().set_facecolor('black')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['release_year'], bins=30, kde=True, color='white')
plt.title('Release Year Distribution')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.gca().set_facecolor('black')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=data, palette='husl')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.gca().set_facecolor('black')
plt.show()
```

```python
top_directors = data['director'].value_counts().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='Paired')
plt.title('Top 10 Directors by Count')
plt.xlabel('Count')
plt.ylabel('Director')
plt.gca().set_facecolor('black')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='country', data=data, order=data['country'].value_counts().index[:10], palette='Set2')
plt.title('Country-wise Distribution (Top 10)')
plt.xlabel('Count')
plt.ylabel('Country')
plt.gca().set_facecolor('black')
plt.show()
```

```python
top_cast = data['cast'].value_counts().nlargest(10)
plt.figure(figsize=(6, 6))
sns.barplot(x=top_cast.values, y=top_cast.index, palette='husl')
plt.title('Top 10 Cast by Count')
plt.xlabel('Count')
plt.ylabel('Cast')
plt.gca().set_facecolor('black')
plt.show()
```

```python
genres = ', '.join(data['listed_in']).split(', ')
plt.figure(figsize=(12, 8))
sns.countplot(y=genres, order=pd.Series(genres).value_counts().index[:10], palette='Set3')
plt.title('List of Genres (Top 10)')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.gca().set_facecolor('black')
plt.show()
```
